```python
Tier_1_Accession = ex_Tier_1.Accession
Tier_3_Accession = ex_Tier_3.Accession
```

```python
#0
fasta.references[0][-8:]
```

```python
need_name_1 = []
need_name_3 = []

for i in range(0,75):
#     print(Tier_1_Accession[i])
    for j in range(0,1490):
        if Tier_1_Accession[i]  == fasta.references[j][-8:]:
            need_name_1.append(fasta.references[j])
            break

for i in range(0,107):
    for j in range(0,1490):
        if Tier_3_Accession[i]  == fasta.references[j][-8:]:
            need_name_3.append(fasta.references[j])
            break
```

```python
Sequence1 = [] 
Sequence3 = [] 

for i in range(0,75):
    Sequence1.append(fasta.fetch(need_name_1[i],0,1059))
for i in range(0,107):
    Sequence3.append(fasta.fetch(need_name_3[i],0,1059))
```

```python
# Sequence
```

```python
data1 = [need_name_1,all_Tier_1,Sequence1]
data3 = [need_name_3,all_Tier_3,Sequence3]

dfData1 = { # DataFrame
        'name':data1[0],
        'Tier':data1[1],
        'Sequence':data1[2]
    }
dfData3 = { 
        'name':data3[0],
        'Tier':data3[1],
        'Sequence':data3[2]
    }

df1 = pd.DataFrame(dfData1) # DataFrame
df3 = pd.DataFrame(dfData3)

# df.to_excel('/www/yang/csv/test.xlsx',index=False) # 0,1,2...
df1.to_excel('/www/yang/csv/end_Tier_1.xlsx',index=False) # 0,1,2...
df3.to_excel('/www/yang/csv/end_Tier_3.xlsx',index=False) # 0,1,2...
```
